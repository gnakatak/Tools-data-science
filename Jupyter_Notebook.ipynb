## Código para mostrar 10 números aleatórios

```python
%matplotlib inline
import random
import matplotlib.pyplot as plt

vet = [random.randint(1, 9999) for i in range(10)]

plt.plot(vet)
plt.show()
```

## Código para o horário em São Paulo

```python
from datetime import datetime
import pytz

fuso_brasilia = pytz.timezone('America/Sao_Paulo')
print(datetime.now(fuso_brasilia))
```

## Código para transformar minutos em horas

```python
min = int(input("digite uma quantidade de minutos"))
print(f"{min//60}:{min%60}")
```

## Multiplicar e somar números

```python
x = int(input("valor de X"))
y = int(input("valor de Y"))        
p = input(f"escolha \n somar: 1 \n diminuir: 2 \n dividir: 3 \n multiplicar: 4 \n")
if p == "1":
    print(x+y)
elif p == "2":
    print(x-y)
elif p == "3":
    print(x/y)
elif p == "4":
    print(x*y)
    
```
